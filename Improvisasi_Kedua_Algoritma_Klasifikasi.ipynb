MENCOBA MEMBANDINGKAN ALGORITMA DECISION TREE DAN K-Nearest Neighbors

```python
# Mengimpor algoritma Decision Tree dari pustaka scikit-learn
from sklearn import tree

# 1. Menyiapkan Data Latih biasa kita sebut Dataset Dummy
# Fitur (X) : Merupakan Data yang kita amati. Formatnya [Nilai UTS, Persentase Kehadiran]
X = [
    [85, 90],  # Mahasiswa 1: UTS 85, Hadir 90%
    [45, 50],  # Mahasiswa 2: UTS 45, Hadir 50%
    [92, 95],  # Mahasiswa 3: UTS 92, Hadir 95%
    [30, 40]   # Mahasiswa 4: UTS 30, Hadir 40%
]

# Label (Y) : Merupakan Target yang ingin diprediksi. 1 untuk Lulus, 0 untuk Tidak Lulus
Y = [1, 0, 1, 0]

# 2. Mengaktifkan dan Melatih Model atau Fase Training
# Membuat baris deklarasi modelnya menjadi Algoritma Decision Tree :
model = tree.DecisionTreeClassifier()
model = model.fit(X, Y) # Menyuruh algoritma belajar mencari pola dari data X dan Y

# 3. Melakukan Prediksi untuk Data Baru (Fase Testing)
# Ada mahasiswa baru yang belum diketahui statusnya.
data_baru = [[70, 80]]
prediksi = model.predict(data_baru)

print("--- Hasil Prediksi Decision Tree ---")
if prediksi[0] == 1:
    print("Sistem memprediksi: Mahasiswa ini kemungkinan LULUS.")
else:
    print("Sistem memprediksi: Mahasiswa ini kemungkinan TIDAK LULUS.")

#==============================================================================

# Mengimpor algoritma K-Nearest Neighbors dari pustaka scikit-learn
from sklearn.neighbors import KNeighborsClassifier

# 1. Menyiapkan Data Latih (Dataset Dummy)
# Fitur (X): Data yang kita amati. Formatnya [Nilai UTS, Persentase Kehadiran]
X = [
    [85, 90],  # Mahasiswa 1: UTS 85, Hadir 90%
    [45, 50],  # Mahasiswa 2: UTS 45, Hadir 50%
    [92, 95],  # Mahasiswa 3: UTS 92, Hadir 95%
    [30, 40]   # Mahasiswa 4: UTS 30, Hadir 40%
]

# Label (Y): Target yang ingin diprediksi. 1 untuk Lulus, 0 untuk Tidak Lulus
Y = [1, 0, 1, 0]

# 2. Mengaktifkan dan Melatih Model (Fase Training)
# Mengubah baris deklarasi modelnya menjadi Algoritma K-Nearest Neighbors :
model = KNeighborsClassifier(n_neighbors=3)
model = model.fit(X, Y) # Menyuruh algoritma belajar mencari pola dari data X dan Y

# 3. Melakukan Prediksi untuk Data Baru (Fase Testing)
data_baru = [[70, 80]]
prediksi = model.predict(data_baru)

print("--- Hasil Prediksi K-Nearest Neighbors ---")
if prediksi[0] == 1:
    print("Sistem memprediksi: Mahasiswa ini kemungkinan LULUS.")
else:
    print("Sistem memprediksi: Mahasiswa ini kemungkinan TIDAK LULUS.")
```

```python
# Mengimpor algoritma Decision Tree dari pustaka scikit-learn
from sklearn import tree

# 1. Menyiapkan Data Latih (Dataset Dummy)
# Fitur (X): Data yang kita amati. Formatnya [Nilai UTS, Persentase Kehadiran]
X = [
    [85, 90],  # Mahasiswa 1: UTS 85, Hadir 90%
    [45, 50],  # Mahasiswa 2: UTS 45, Hadir 50%
    [92, 95],  # Mahasiswa 3: UTS 92, Hadir 95%
    [30, 40]   # Mahasiswa 4: UTS 30, Hadir 40%
]

# Label (Y): Target yang ingin diprediksi. 1 untuk Lulus, 0 untuk Tidak Lulus
Y = [1, 0, 1, 0]

# 2. Mengaktifkan dan Melatih Model (Fase Training)
# Membuat baris deklarasi modelnya menjadi Algoritma Decision Tree :
model = tree.DecisionTreeClassifier()
model = model.fit(X, Y) # Menyuruh algoritma belajar mencari pola dari data X dan Y

# 3. Melakukan Prediksi untuk Data Baru (Fase Testing)
# Uji Coba Data Nilai Mahasiswa Ekstrem
data_baru = [[40, 45]]
prediksi = model.predict(data_baru)

print("--- Hasil Prediksi Decision Tree ---")
if prediksi[0] == 1:
    print("Sistem memprediksi: Mahasiswa ini kemungkinan LULUS.")
else:
    print("Sistem memprediksi: Mahasiswa ini kemungkinan TIDAK LULUS.")

#==============================================================================

# Mengimpor algoritma K-Nearest Neighbors dari pustaka scikit-learn
from sklearn.neighbors import KNeighborsClassifier

# 1. Menyiapkan Data Latih (Dataset Dummy)
# Fitur (X): Data yang kita amati. Formatnya [Nilai UTS, Persentase Kehadiran]
X = [
    [85, 90],  # Mahasiswa 1: UTS 85, Hadir 90%
    [45, 50],  # Mahasiswa 2: UTS 45, Hadir 50%
    [92, 95],  # Mahasiswa 3: UTS 92, Hadir 95%
    [30, 40]   # Mahasiswa 4: UTS 30, Hadir 40%
]

# Label (Y): Target yang ingin diprediksi. 1 untuk Lulus, 0 untuk Tidak Lulus
Y = [1, 0, 1, 0]

# 2. Mengaktifkan dan Melatih Model (Fase Training)
# Mengubah baris deklarasi modelnya menjadi Algoritma K-Nearest Neighbors :
model = KNeighborsClassifier(n_neighbors=3)
model = model.fit(X, Y) # Menyuruh algoritma belajar mencari pola dari data X dan Y


# 3. Melakukan Prediksi untuk Data Baru (Fase Testing)
# Uji Coba Data Nilai Mahasiswa Ekstrem
data_baru = [[40, 45]]
prediksi = model.predict(data_baru)

print("--- Hasil Prediksi K-Nearest Neighbors ---")
if prediksi[0] == 1:
    print("Sistem memprediksi: Mahasiswa ini kemungkinan LULUS.")
else:
    print("Sistem memprediksi: Mahasiswa ini kemungkinan TIDAK LULUS.")
```
